```python
def get_measures(df, task, condition, column):
    min_ = df.min()
    max_ = df.max()
    mean = df.mean()
    median = df.median()
    std = df.std()
    
    print(f'{condition} {column} for {task}')
    print(f'Min: {min_} - Max: {max_} Mean: {mean} - Median: {median} - STD:{std}\n')
```

```python
# Post-questionnaire analysis

import numpy as np
import matplotlib.pyplot as plt

questions = ['Q1 - I have found the provided user stories:', 
             'Q2 - I have found the development task:', 
             'Q3 - Applying % has been:']

category_names = ['Very  hard', 'Hard', 'Neither easy nor hard', 'Easy', 'Very easy']

def get_percentages(input_list):
    return [i / sum(input_list) * 100 for i in input_list]

def get_values(df, condition):
    
    tot_a1 = [0, 0, 0, 0, 0]
    tot_a2 = [0, 0, 0, 0, 0]
    tot_a3 = [0, 0, 0, 0, 0]
    
    a1 = df.loc[:, 'Q1'].tolist()
    a2 = df.loc[:, 'Q2'].tolist()
    a3 = df.loc[:, 'Q3'].tolist()
    
    for answers, tot_answers in zip([a1, a2, a3], [tot_a1, tot_a2, tot_a3]):  
        for answ in answers:
            tot_answers[answ - 1] += 1
    
    q3 = questions[2].replace('%', 'TDD') if condition == 'TDD' else \
        questions[2].replace('%', 'NO-TDD')

    results = {
        questions[0] : get_percentages(tot_a1),
        questions[1] : get_percentages(tot_a2),
        q3: get_percentages(tot_a3)
    }
    return results
    
def plot_stacked_bars(results, category_names):  
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    middle_index = data.shape[1]//2
    offsets = data[:, range(middle_index)].sum(axis=1) + data[:, middle_index]/2
    
    # Color Mapping
    category_colors = plt.get_cmap('coolwarm_r')(
        np.linspace(0.15, 0.85, data.shape[1]))
    
    fig, ax = plt.subplots(figsize=(12, 10))
    
    # Plot Bars
    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths - offsets
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)
    
    # Add Zero Reference Line
    ax.axvline(0, linestyle='--', color='black', alpha=.25)
    
    # X Axis
    ax.set_xlim(-100, 100)
    ax.set_xticks(np.arange(-100, 100, 10))
    ax.xaxis.set_major_formatter(lambda x, pos: str(abs(int(x))))
    
    # Y Axis
    ax.invert_yaxis()
    
    # Remove spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    # Legend
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1), loc='lower left', fontsize='small')
    
    # Set Background Color
    fig.set_facecolor('#FFFFFF')

    return fig, ax
```

```python
results = get_values(p1_df.loc[p1_df['Condition'] == 'TDD'], 'TDD')  
fig, ax = plot_stacked_bars(results, category_names)
plt.savefig('diverging_stacked_bars_charts/task1_tdd', bbox_inches = 'tight')
    
results = get_values(p1_df.loc[p1_df['Condition'] == 'NO-TDD'], 'NO-TDD')  
fig, ax = plot_stacked_bars(results, category_names)
plt.savefig('diverging_stacked_bars_charts/task1_notdd', bbox_inches = 'tight')
    
results = get_values(p2_df.loc[p2_df['Condition'] == 'TDD'], 'TDD')  
fig, ax = plot_stacked_bars(results, category_names)
plt.savefig('diverging_stacked_bars_charts/task2_tdd', bbox_inches = 'tight')
    
results = get_values(p2_df.loc[p2_df['Condition'] == 'NO-TDD'], 'NO-TDD')
fig, ax = plot_stacked_bars(results, category_names)
plt.savefig('diverging_stacked_bars_charts/task2_notdd', bbox_inches = 'tight')
```
