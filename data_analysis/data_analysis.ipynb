```python
p1_df = df12.loc[df12['Period'] == 'P1']
p2_df = df12.loc[df12['Period'] == 'P2']
```

```python
# Boxplot for QLTY in task 1 between TDD and NO-TDD
qlty_boxplot_p1 = p1_df.boxplot(by='Condition', column=['QLTY'], grid=True)
```

```python
p1_tdd_qlty_df = p1_df.loc[p1_df['Condition'] == 'TDD']['QLTY']
p1_tdd_qlty_min = p1_tdd_qlty_df.min()
p1_tdd_qlty_max = p1_tdd_qlty_df.max()
p1_tdd_qlty_mean = p1_tdd_qlty_df.mean()
p1_tdd_qlty_median = p1_tdd_qlty_df.median()
p1_tdd_qlty_std = p1_tdd_qlty_df.std()

print('TDD')
print(f'Min: {p1_tdd_qlty_min} - Max: {p1_tdd_qlty_max}')
print(f'Mean: {p1_tdd_qlty_mean} - Median: {p1_tdd_qlty_median} - STD:{p1_tdd_qlty_std}\n')

p1_notdd_qlty_df = p1_df.loc[p1_df['Condition'] == 'NO-TDD']['QLTY']
p1_notdd_qlty_min = p1_notdd_qlty_df.min()
p1_notdd_qlty_max = p1_notdd_qlty_df.max()
p1_notdd_qlty_mean = p1_notdd_qlty_df.mean()
p1_notdd_qlty_median = p1_notdd_qlty_df.median()
p1_notdd_qlty_std = p1_notdd_qlty_df.std()

print('NO-TDD')
print(f'Min: {p1_notdd_qlty_min} - Max: {p1_notdd_qlty_max}')
print(f'Mean: {p1_notdd_qlty_mean} - Median: {p1_notdd_qlty_median} - STD:{p1_notdd_qlty_std}')
```
