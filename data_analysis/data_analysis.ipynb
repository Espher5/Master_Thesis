```python
# Post-questionnaire analysis
import numpy as np
import matplotlib.pyplot as plt

questions = ['Q1 ', 'Q2 ', 'Q3 (%)']
category_names = ['Very hard', 'Hard', 'Neither easy nor hard', 'Easy', 'Very easy']

def get_percentages(input_list):
    return [i / sum(input_list) * 100 for i in input_list]

def get_values(df, condition):
    tot_a1 = [0, 0, 0, 0, 0]
    tot_a2 = [0, 0, 0, 0, 0]
    tot_a3 = [0, 0, 0, 0, 0]
    
    a1 = df.loc[:, 'Q1'].tolist()
    a2 = df.loc[:, 'Q2'].tolist()
    a3 = df.loc[:, 'Q3'].tolist()
    
    for answers, tot_answers in zip([a1, a2, a3], [tot_a1, tot_a2, tot_a3]):  
        for answ in answers:
            tot_answers[answ - 1] += 1
    
    q1 = questions[0] + '(' + condition + ')' 
    q2 = questions[1] + '(' + condition + ')'
    q3 = questions[2].replace('%', 'TDD') if condition == 'TDD' else \
        questions[2].replace('%', 'NO-TDD')

    results = [
        (q1, get_percentages(tot_a1)),
        (q2, get_percentages(tot_a2)),
        (q3, get_percentages(tot_a3))
    ]
    return results
    
def plot_stacked_bars(results):  
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    middle_index = data.shape[1]//2
    offsets = data[:, range(middle_index)].sum(axis=1) + data[:, middle_index]/2
    
    # Color Mapping
    category_colors = plt.get_cmap('coolwarm_r')(np.linspace(0.15, 0.85, data.shape[1]))
    fig, ax = plt.subplots(figsize=(12, 2))
    
    # Plot Bars
    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        start = data_cum[:, i] - widths - offsets
        ax.barh(y=labels, width=widths, height=0.2, left=start, label=colname, color=color)

    
    # Add Zero Reference Line
    ax.axvline(0, linestyle='--', color='black', alpha=.25)
    
    # X Axis
    ax.set_xlim(-100, 100)
    ax.set_xticks(np.arange(-100, 100, 10))
    ax.xaxis.set_major_formatter(lambda x, pos: str(abs(int(x))))
    
    # Y Axis
    ax.invert_yaxis()

    # Remove spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    # Legend
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1), loc='lower left')
    
    # Set Background Color
    fig.set_facecolor('#FFFFFF')

    return fig, ax
```

```python
# Sorry for this monstrosity :(
def lists_to_dict(dict1, dict2):
    res_dict = {}
    res_dict.update({dict1[0][0]: dict1[0][1]})
    res_dict.update({dict2[0][0]: dict2[0][1]})
    res_dict.update({dict1[1][0]: dict1[1][1]})
    res_dict.update({dict2[1][0]: dict2[1][1]})
    res_dict.update({dict1[2][0]: dict1[2][1]})
    res_dict.update({dict2[2][0]: dict2[2][1]})
    
    return res_dict

def draw_bar_charts():
    for task_df, task in zip([t1_df, t2_df], ['task1', 'task2']):
        results_tdd = get_values(task_df.loc[task_df['Condition'] == 'TDD'], 'TDD')
        results_dict_tdd = {} 
        for (q, a) in results_tdd:
            results_dict_tdd.update({q: a})
        fig, ax = plot_stacked_bars(results_dict_tdd)
        plt.savefig('bar_charts/' + task + '_tdd', bbox_inches = 'tight')
            
        results_notdd = get_values(task_df.loc[task_df['Condition'] == 'NO-TDD'], 'NO-TDD')
        results_dict_notdd = {}
        for (q, a) in results_notdd:
             results_dict_notdd.update({q: a})
        fig, ax = plot_stacked_bars(results_dict_notdd)
        plt.savefig('bar_charts/' + task + '_notdd', bbox_inches = 'tight')
           
        res_dict = lists_to_dict(results_tdd, results_notdd)
        fig, ax = plot_stacked_bars(res_dict)
        plt.savefig('bar_charts/' + task, bbox_inches = 'tight')
```

```python
# Meta-analysis on both studies
import PythonMeta as PMA

def showstudies(studies,dtype):    
    #show continuous data
    if dtype.upper()=="CONT":
        text = "%-10s %-30s %-30s \n"%("Study ID","Experiment Group","Control Group")
        text += "%-10s %-10s %-10s %-10s %-10s %-10s %-10s \n"%(" ","m1","sd1","n1","m2","sd2","n2")
        for i in range(len(studies)):
            text += "%-10s %-10s %-10s %-10s %-10s  %-10s %-10s \n"%(
            studies[i][6],        #study ID
            str(studies[i][0]),   #mean of group1
            str(studies[i][1]),   #SD of group1
            str(studies[i][2]),   #total num of group1
            str(studies[i][3]),   #mean of group2
            str(studies[i][4]),   #SD of group2
            str(studies[i][5])    #total num of group2
            )
        return text
        
    #show dichotomous data
    text = "%-10s %-20s %-20s \n"%("Study ID","Experiment Group","Control Group")
    text += "%-10s %-10s %-10s %-10s %-10s \n"%(" ","e1","n1","e2","n2")
    for i in range(len(studies)):
        text += "%-10s %-10s %-10s %-10s %-10s \n"%(
        studies[i][4],        #study ID
        str(studies[i][0]),   #event num of group1
        str(studies[i][1]),   #total num of group1
        str(studies[i][2]),   #event num of group2
        str(studies[i][3])    #total num of group2
        )
    return text

def showresults(rults):
    text = "%-10s %-6s  %-18s %-10s"%("Study ID","n","ES[95% CI]","Weight(%)\n")    
    for i in range(1,len(rults)):
        text += "%-10s %-6d  %-4.2f[%.2f %.2f]   %6.2f\n"%(   # for each study
        rults[i][0],     #study ID
        rults[i][5],     #total num
        rults[i][1],     #effect size
        rults[i][3],     #lower of CI
        rults[i][4],     #higher of CI
        100*(rults[i][2]/rults[0][2])  #weight
        )
    text += "%-10s %-6d  %-4.2f[%.2f %.2f]   %6d\n"%(         # for total effect
        rults[0][0],     #total effect size name
        rults[0][5],     #total N (all studies)
        rults[0][1],     #total effect size
        rults[0][3],     #total lower CI
        rults[0][4],     #total higher CI
        100
        )  
    text += "%d studies included (N=%d)\n"%(len(rults)-1,rults[0][5])
    text += "Heterogeneity: Tau\u00b2=%.3f "%(rults[0][12]) if not rults[0][12]==None else "Heterogeneity: "
    text += "Q(Chisquare)=%.2f(p=%s); I\u00b2=%s\n"%(
        rults[0][7],     #Q test value
        rults[0][8],     #p value for Q test
        str(round(rults[0][9],2))+"%")   #I-square value
    text += "Overall effect test: z=%.2f, p=%s\n"%(rults[0][10],rults[0][11])  #z-test value and p-value
    
    return text

def main(stys, settings, variable):
    d = PMA.Data()  #Load Data class
    m = PMA.Meta()  #Load Meta class
    f = PMA.Fig()   #Load Fig class
    
    with open(variable + '.txt', 'w') as out_file:
        #You should always tell the datatype first!!!
        d.datatype = settings["datatype"]                #set data type, 'CATE' for binary data or 'CONT' for continuous data
        studies = d.getdata(stys)                        #load data
        #studies = d.getdata(d.readfile("studies.txt"))  #get data from a data file, see examples of data files
        print(showstudies(studies,d.datatype))           
        out_file.write(showstudies(studies,d.datatype))

        m.subgroup=d.subgroup                            #set the subgroup
        m.datatype=d.datatype                            #set data type for meta-analysis calculating
        m.models = settings["models"]                    #set effect models: 'Fixed' or 'Random'
        m.algorithm = settings["algorithm"]              #set algorithm, based on datatype and effect size
        m.effect = settings["effect"]                    #set effect size:RR/OR/RD for binary data; SMD/MD for continuous data
        results = m.meta(studies)                        #performing the analysis
        print(m.models + " " + m.algorithm + " " + m.effect)
        print (showresults(results))                     #show results table
        out_file.write(m.models + " " + m.algorithm + " " + m.effect)
        out_file.write(showresults(results))
        
        f.forest(results).show()                         #show forest plot
        f.funnel(results).show()                         #show funnel plot
        print(m.Eggers_test(results))                    #Eggers_test: (Intercept, t-value, p-value, SD  and CI)
        out_file.write(str(m.Eggers_test(results)))
    
```
